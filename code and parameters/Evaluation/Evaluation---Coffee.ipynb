```python
##################################################################################################
# Author：Yun Zheng
# Email：18848234471@163.com
# If there are any shortcomings, please forgive me,because this is my first time sharing.
##################################################################################################
# Data Augmentation - Coffee
# package
import numpy as np
import pandas as pd
from skimage.metrics import structural_similarity as ssim
import matplotlib.pyplot as plt
from scipy.spatial.distance import euclidean

##加载原始数据集
df = pd.read_excel('coffee.xlsx', header=None)
coffee1 = df.iloc[1:11, 1:].to_numpy()
coffee2 = df.iloc[11:21, 1:].to_numpy()
coffee3 = df.iloc[21:31, 1:].to_numpy()
coffee4 = df.iloc[31:41, 1:].to_numpy()
coffee5 = df.iloc[41:51, 1:].to_numpy()
coffee6 = df.iloc[51:61, 1:].to_numpy()
coffee7 = df.iloc[61:71, 1:].to_numpy()
coffee = np.concatenate((coffee1, coffee2, coffee3, coffee4, coffee5, coffee6, coffee7))
##加载每类Bootstrap生成数据集
df = pd.read_csv('1400-coffee-B.csv', header=None)
coffee_B_1 = df.iloc[0:200, 0:].to_numpy()
coffee_B_2 = df.iloc[200:400, 0:].to_numpy()
coffee_B_3 = df.iloc[400:600, 0:].to_numpy()
coffee_B_4 = df.iloc[600:800, 0:].to_numpy()
coffee_B_5 = df.iloc[800:1000, 0:].to_numpy()
coffee_B_6 = df.iloc[1000:1200, 0:].to_numpy()
coffee_B_7 = df.iloc[1200:1400, 0:].to_numpy()
coffee_B = df.iloc[0:1400, 0:].to_numpy()
##加载每类GAN生成数据集
df = pd.read_csv('1400_G_coffee.csv', header=None)
coffee_G_1 = df.iloc[0:200, 0:].to_numpy()
coffee_G_2 = df.iloc[200:400, 0:].to_numpy()
coffee_G_3 = df.iloc[400:600, 0:].to_numpy()
coffee_G_4 = df.iloc[600:800, 0:].to_numpy()
coffee_G_5 = df.iloc[800:1000, 0:].to_numpy()
coffee_G_6 = df.iloc[1000:1200, 0:].to_numpy()
coffee_G_7 = df.iloc[1200:1400, 0:].to_numpy()
coffee_G = df.iloc[0:1400, 0:].to_numpy()
##合并B-G形式的数据集
coffee_B_G_1 = np.concatenate((coffee_B_1, coffee_G_1))
coffee_B_G_2 = np.concatenate((coffee_B_2, coffee_G_2))
coffee_B_G_3 = np.concatenate((coffee_B_3, coffee_G_3))
coffee_B_G_4 = np.concatenate((coffee_B_4, coffee_G_4))
coffee_B_G_5 = np.concatenate((coffee_B_5, coffee_G_5))
coffee_B_G_6 = np.concatenate((coffee_B_6, coffee_G_6))
coffee_B_G_7 = np.concatenate((coffee_B_7, coffee_G_7))
coffee_B_G = np.concatenate((coffee_B_G_1, coffee_B_G_2, coffee_B_G_3, coffee_B_G_4, coffee_B_G_5, coffee_B_G_6, coffee_B_G_7))
##合并Y-B-G形式的数据集
coffee_Y_B_G_1 = np.concatenate((coffee1, coffee_B_1, coffee_G_1))
coffee_Y_B_G_2 = np.concatenate((coffee2, coffee_B_2, coffee_G_2))
coffee_Y_B_G_3 = np.concatenate((coffee3, coffee_B_3, coffee_G_3))
coffee_Y_B_G_4 = np.concatenate((coffee4, coffee_B_4, coffee_G_4))
coffee_Y_B_G_5 = np.concatenate((coffee5, coffee_B_5, coffee_G_5))
coffee_Y_B_G_6 = np.concatenate((coffee6, coffee_B_6, coffee_G_6))
coffee_Y_B_G_7 = np.concatenate((coffee7, coffee_B_7, coffee_G_7))
coffee_Y_B_G = np.concatenate((coffee_Y_B_G_1, coffee_Y_B_G_2, coffee_Y_B_G_3, coffee_Y_B_G_4, coffee_Y_B_G_5, coffee_Y_B_G_6, coffee_Y_B_G_7))
###Y-B
df = pd.read_csv('1470-coffee-Y-B.csv', header=None)
coffee_Y_B = df.iloc[1:1470, 1:].to_numpy()
###Y-G
df = pd.read_csv('1470_Y_G_coffee.csv', header=None)
coffee_Y_G = df.iloc[0:1470, 0:].to_numpy()
###############################################################################################
###############################################################################################
####计算Bootstrap的评价指标
avg_yaopian = np.mean(coffee, axis=0)
avg_yaopian_B = np.mean(coffee_B, axis=0)
# 假设Z和S分别是形状为（70，601）的numpy数组
correlation = np.corrcoef(avg_yaopian, avg_yaopian_B)[0, 1]
print(correlation)
#####平均光谱SSIM值
# 计算Z和S的SSIM值
ssim_value, _ = ssim(avg_yaopian, avg_yaopian_B, full=True, data_range=np.max(avg_yaopian) - np.min(avg_yaopian))
print(ssim_value)
# 计算切比雪夫距离
chebyshev_distance = np.abs(avg_yaopian - avg_yaopian_B).max()
print(chebyshev_distance)

####计算GAN的评价指标
avg_yaopian = np.mean(coffee, axis=0)
avg_yaopian_G = np.mean(coffee_G, axis=0)
# 假设Z和S分别是形状为（70，601）的numpy数组
correlation = np.corrcoef(avg_yaopian, avg_yaopian_G)[0, 1]
print(correlation)
#####平均光谱SSIM值
# 计算Z和S的SSIM值
ssim_value, _ = ssim(avg_yaopian, avg_yaopian_G, full=True, data_range=np.max(avg_yaopian) - np.min(avg_yaopian))
print(ssim_value)
# 计算切比雪夫距离
chebyshev_distance = np.abs(avg_yaopian - avg_yaopian_G).max()
print(chebyshev_distance)

####计算Bootstrap---GAN的评价指标
avg_yaopian = np.mean(coffee, axis=0)
avg_yaopian_B_G = np.mean(coffee_B_G, axis=0)
# 假设Z和S分别是形状为（70，601）的numpy数组
correlation = np.corrcoef(avg_yaopian, avg_yaopian_B_G)[0, 1]
print(correlation)
#####平均光谱SSIM值
# 计算Z和S的SSIM值
ssim_value, _ = ssim(avg_yaopian, avg_yaopian_B_G, full=True, data_range=np.max(avg_yaopian) - np.min(avg_yaopian))
print(ssim_value)
# 计算切比雪夫距离
chebyshev_distance = np.abs(avg_yaopian - avg_yaopian_B_G).max()
print(chebyshev_distance)

####计算Y--B的评价指标
avg_yaopian = np.mean(coffee, axis=0)
avg_yaopian_Y_B = np.mean(coffee_Y_B, axis=0)
# 假设Z和S分别是形状为（70，601）的numpy数组
correlation = np.corrcoef(avg_yaopian, avg_yaopian_Y_B)[0, 1]
print(correlation)
#####平均光谱SSIM值
# 计算Z和S的SSIM值
ssim_value, _ = ssim(avg_yaopian, avg_yaopian_Y_B, full=True, data_range=np.max(avg_yaopian) - np.min(avg_yaopian))
print(ssim_value)
# 计算切比雪夫距离
chebyshev_distance = np.abs(avg_yaopian - avg_yaopian_Y_B).max()
print(chebyshev_distance)

####计算Y--G的评价指标
avg_yaopian = np.mean(coffee, axis=0)
avg_yaopian_Y_G = np.mean(coffee_Y_G, axis=0)
# 假设Z和S分别是形状为（70，601）的numpy数组
correlation = np.corrcoef(avg_yaopian, avg_yaopian_Y_G)[0, 1]
print(correlation)
#####平均光谱SSIM值
# 计算Z和S的SSIM值
ssim_value, _ = ssim(avg_yaopian, avg_yaopian_Y_G, full=True, data_range=np.max(avg_yaopian) - np.min(avg_yaopian))
print(ssim_value)
# 计算切比雪夫距离
chebyshev_distance = np.abs(avg_yaopian - avg_yaopian_Y_G).max()
print(chebyshev_distance)

####计算Y-B-G的评价指标
avg_yaopian = np.mean(coffee, axis=0)
avg_yaopian_Y_B_G = np.mean(coffee_Y_B_G, axis=0)
# 假设Z和S分别是形状为（70，601）的numpy数组
correlation = np.corrcoef(avg_yaopian, avg_yaopian_Y_B_G)[0, 1]
print(correlation)
#####平均光谱SSIM值
# 计算Z和S的SSIM值
ssim_value, _ = ssim(avg_yaopian, avg_yaopian_Y_B_G, full=True, data_range=np.max(avg_yaopian) - np.min(avg_yaopian))
print(ssim_value)
# 计算切比雪夫距离
chebyshev_distance = np.abs(avg_yaopian - avg_yaopian_Y_B_G).max()
print(chebyshev_distance)
```
