```python
##################################################################################################
# Author：Yun Zheng
# Email：18848234471@163.com
# If there are any shortcomings, please forgive me,because this is my first time sharing.
##################################################################################################
# Model - SVM--Tablet - matlab(Y、Y-B、Y-B-G)
# It can also be run with the same parameters in Python, but we believe Matlab would be more intuitive
%%  清空环境变量
warning off             % 关闭报警信息
close all               % 关闭开启的图窗
clear                   % 清空变量
clc                     % 清空命令行

%%  导入数据
filename = 'D:\tablet-3-4.csv';
data = csvread(filename);
x = data(2:end,2:end);
y = data(2:end,1);

% 存储30次的error1和error2
errors1 = zeros(30, 1);
errors2 = zeros(30, 1);
for i = 1:30
    %%  KS法划分训练集和测试集
    ratio=0.8;
    [mx,nx]=size(x);
    mtrain=ceil(mx*ratio);
    mtest=mx-mtrain;
    c = cvpartition(y, 'HoldOut', 0.2); % 划分成训练集和测试集，测试集占比 20%
    trainIdx = training(c);
    testIdx = test(c);
    P_train = x(trainIdx, :); % 训练集特征数据
    T_train = y(trainIdx, :); % 训练集标签数据
    P_test = x(testIdx, :); % 测试集特征数据
    T_test = y(testIdx, :); % 测试集标签数据
    P_train = P_train';
    P_test = P_test';
    T_train = T_train';
    T_test = T_test';
    M = size(P_train, 2);
    N = size(P_test, 2);
    
    %%  数据归一化
    [p_train, ps_input] = mapminmax(P_train, 0, 1);
    p_test = mapminmax('apply', P_test, ps_input );
    t_train = T_train;
    t_test  = T_test ;
    
    %%  转置以适应模型
    p_train = p_train'; p_test = p_test';
    t_train = t_train'; t_test = t_test';
    
    %%  创建模型
    c = 1;      % 惩罚因子
    cmd = ['-t 0', '-c', num2str(c), '-g', num2str(g)];
    model = svmtrain(t_train, p_train, cmd);
    
    %%  仿真测试
    T_sim1 = svmpredict(t_train, p_train, model);
    T_sim2 = svmpredict(t_test , p_test , model);
    
    %%  性能评价
    errors1(i) = sum((T_sim1' == T_train)) / M * 100 ;
    errors2(i) = sum((T_sim2' == T_test )) / N * 100 ;
end

% 打印30次的error1和error2
disp('30次的error1和error2：');
disp('error1:');
disp(errors1);
disp('error2:');
disp(errors2);

% 计算30次的平均值
avg_error1 = mean(errors1);
avg_error2 = mean(errors2);

disp(['30次的平均error1: ', num2str(avg_error1)]);
disp(['30次的平均error2: ', num2str(avg_error2)])
##################################################################################################
# Author：Yun Zheng
# Email：18848234471@163.com
# If there are any shortcomings, please forgive me,because this is my first time sharing.
##################################################################################################
# Model - SVM--Coffee - matlab(Y、Y-B、Y-B-G)
# It can also be run with the same parameters in Python, but we believe Matlab would be more intuitive
%%  清空环境变量
warning off             % 关闭报警信息
close all               % 关闭开启的图窗
clear                   % 清空变量
clc                     % 清空命令行

%%  导入数据
filename = 'D:\coffee.xlsx';
data = xlsread(filename);
x = data(2:end,2:end);
y = data(2:end,1);

% 存储30次的error1和error2
errors1 = zeros(30, 1);
errors2 = zeros(30, 1);
for i = 1:30
    %%  KS法划分训练集和测试集
    ratio=0.8;
    [mx,nx]=size(x);
    mtrain=ceil(mx*ratio);
    mtest=mx-mtrain;
    c = cvpartition(y, 'HoldOut', 0.2); % 划分成训练集和测试集，测试集占比 20%
    trainIdx = training(c);
    testIdx = test(c);
    P_train = x(trainIdx, :); % 训练集特征数据
    T_train = y(trainIdx, :); % 训练集标签数据
    P_test = x(testIdx, :); % 测试集特征数据
    T_test = y(testIdx, :); % 测试集标签数据
    P_train = P_train';
    P_test = P_test';
    T_train = T_train';
    T_test = T_test';
    M = size(P_train, 2);
    N = size(P_test, 2);
    
    %%  数据归一化
    [p_train, ps_input] = mapminmax(P_train, 0, 1);
    p_test = mapminmax('apply', P_test, ps_input );
    t_train = T_train;
    t_test  = T_test ;
    
    %%  转置以适应模型
    p_train = p_train'; p_test = p_test';
    t_train = t_train'; t_test = t_test';
    
    %%  创建模型
    c = 1;      % 惩罚因子
    cmd = ['-t 0', '-c', num2str(c), '-g', num2str(g)];
    model = svmtrain(t_train, p_train, cmd);
    
    %%  仿真测试
    T_sim1 = svmpredict(t_train, p_train, model);
    T_sim2 = svmpredict(t_test , p_test , model);
    
    %%  性能评价
    errors1(i) = sum((T_sim1' == T_train)) / M * 100 ;
    errors2(i) = sum((T_sim2' == T_test )) / N * 100 ;
end

% 打印30次的error1和error2
disp('30次的error1和error2：');
disp('error1:');
disp(errors1);
disp('error2:');
disp(errors2);

% 计算30次的平均值
avg_error1 = mean(errors1);
avg_error2 = mean(errors2);

disp(['30次的平均error1: ', num2str(avg_error1)]);
disp(['30次的平均error2: ', num2str(avg_error2)])
##################################################################################################
# Author：Yun Zheng
# Email：18848234471@163.com
# If there are any shortcomings, please forgive me,because this is my first time sharing.
##################################################################################################
# Model - SVM--Grape - matlab(Y、Y-B、Y-B-G)
# It can also be run with the same parameters in Python, but we believe Matlab would be more intuitive
%%  清空环境变量
warning off             % 关闭报警信息
close all               % 关闭开启的图窗
clear                   % 清空变量
clc                     % 清空命令行

%%  导入数据
filename = 'D:\putao.xlsx';
data = xlsread(filename);
x = data(2:end,2:end);
y = data(2:end,1);

% 存储30次的error1和error2
errors1 = zeros(30, 1);
errors2 = zeros(30, 1);
for i = 1:30
    %%  KS法划分训练集和测试集
    ratio=0.8;
    [mx,nx]=size(x);
    mtrain=ceil(mx*ratio);
    mtest=mx-mtrain;
    c = cvpartition(y, 'HoldOut', 0.2); % 划分成训练集和测试集，测试集占比 20%
    trainIdx = training(c);
    testIdx = test(c);
    P_train = x(trainIdx, :); % 训练集特征数据
    T_train = y(trainIdx, :); % 训练集标签数据
    P_test = x(testIdx, :); % 测试集特征数据
    T_test = y(testIdx, :); % 测试集标签数据
    P_train = P_train';
    P_test = P_test';
    T_train = T_train';
    T_test = T_test';
    M = size(P_train, 2);
    N = size(P_test, 2);
    
    %%  数据归一化
    [p_train, ps_input] = mapminmax(P_train, 0, 1);
    p_test = mapminmax('apply', P_test, ps_input );
    t_train = T_train;
    t_test  = T_test ;
    
    %%  转置以适应模型
    p_train = p_train'; p_test = p_test';
    t_train = t_train'; t_test = t_test';
    
    %%  创建模型
    c = 1;      % 惩罚因子
    cmd = ['-t 0', '-c', num2str(c), '-g', num2str(g)];
    model = svmtrain(t_train, p_train, cmd);
    
    %%  仿真测试
    T_sim1 = svmpredict(t_train, p_train, model);
    T_sim2 = svmpredict(t_test , p_test , model);
    
    %%  性能评价
    errors1(i) = sum((T_sim1' == T_train)) / M * 100 ;
    errors2(i) = sum((T_sim2' == T_test )) / N * 100 ;
end

% 打印30次的error1和error2
disp('30次的error1和error2：');
disp('error1:');
disp(errors1);
disp('error2:');
disp(errors2);

% 计算30次的平均值
avg_error1 = mean(errors1);
avg_error2 = mean(errors2);

disp(['30次的平均error1: ', num2str(avg_error1)]);
disp(['30次的平均error2: ', num2str(avg_error2)])
```
