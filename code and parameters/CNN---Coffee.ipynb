```python
##################################################################################################
# Author：Yun Zheng
# Email：18848234471@163.com
# If there are any shortcomings, please forgive me,because this is my first time sharing.
##################################################################################################
# Model - CNN--Tablet - matlab(Y、Y-B、Y-B-G)
# It can also be run with the same parameters in Python, but we believe Matlab would be more intuitive
%%  清空环境变量
warning off             % 关闭报警信息
close all               % 关闭开启的图窗
clear                   % 清空变量
clc                     % 清空命令行
%%  导入数据
filename = 'D:\yaopian\Y-B\3360_Y_B_yaopin.csv';
data = csvread(filename);
x = data(2:end,2:end);
y = data(2:end,1);
% 存储30次的error1和error2
errors1 = zeros(30, 1)
errors2 = zeros(30, 1);
for i = 1:30
    %%  KS法划分训练集和测试集
    ratio=0.8;
    [mx,nx]=size(x);
    mtrain=ceil(mx*ratio);
    mtest=mx-mtrain;
    c = cvpartition(y, 'HoldOut', 0.2);
    trainIdx = training(c);
    testIdx = test(c);
    P_train = x(trainIdx, :); 
    T_train = y(trainIdx, :); 
    P_test = x(testIdx, :); 
    T_test = y(testIdx, :); 
    P_train = P_train';
    P_test = P_test';
    T_train = T_train';
    T_test = T_test';
    M = size(P_train, 2);
    N = size(P_test, 2);
    
    %%  数据归一化
    [P_train, ps_input] = mapminmax(P_train, 0, 1);
    P_test  = mapminmax('apply', P_test, ps_input);
    
    t_train =  categorical(T_train)';
    t_test  =  categorical(T_test )';
    
    %%  数据平铺
    p_train =  double(reshape(P_train, 404, 1, 1, M)); 
    p_test  =  double(reshape(P_test , 404, 1, 1, N));
    
    %%  构造网络结构
    layers = [
     imageInputLayer([404,1,1])             
    
     convolution2dLayer([4,1], 16,'Stride' , [1,1], 'Padding','same')         
     batchNormalizationLayer                 
     reluLayer                              
     
     convolution2dLayer([4,1], 32,'Stride' , [1,1], 'Padding','same')          
     batchNormalizationLayer                
     reluLayer                               
     maxPooling2dLayer([4,1], 'Stride' , [2,1])  
    
     fullyConnectedLayer(2)                  
     softmaxLayer                           
     classificationLayer];                  
    
    %%  参数设置
    options = trainingOptions('adam', ...      
        'MaxEpochs', 50, ...                  
        'MiniBatchSize', 40, ...                  
        'InitialLearnRate', 1e-4, ...         
        'L2Regularization', 1e-2, ...         
        'LearnRateSchedule', 'piecewise', ...  
        'LearnRateDropFactor', 0.5, ...       
        'LearnRateDropPeriod', 50, ...        
        'Shuffle', 'every-epoch', ...          
        'ValidationPatience', Inf, ...         
        'Plots', 'training-progress', ...      
        'Verbose', false);
    
    
    %%  训练模型
    net = trainNetwork(p_train, t_train, layers, options);
    
    %%  预测模型
    t_sim1 = predict(net, p_train); 
    t_sim2 = predict(net, p_test ); 
    
    %%  反归一化
    T_sim1 = vec2ind(t_sim1');
    T_sim2 = vec2ind(t_sim2');
    
    %%  性能评价
    errors1(i) = sum((T_sim1 == T_train)) / M * 100 ;
    errors2(i) = sum((T_sim2 == T_test )) / N * 100 ;
end

% 打印30次的error1和error2
disp('30次的error1和error2：');
disp('error1:');
disp(errors1);
disp('error2:');
disp(errors2);

% 计算30次的平均值
avg_error1 = mean(errors1);
avg_error2 = mean(errors2);

disp(['30次的平均error1: ', num2str(avg_error1)]);
disp(['30次的平均error2: ', num2str(avg_error2)])

##################################################################################################
# Author：Yun Zheng
# Email：18848234471@163.com
# If there are any shortcomings, please forgive me,because this is my first time sharing.
##################################################################################################
# Model - CNN--Coffee - matlab(Y、Y-B、Y-B-G)
# It can also be run with the same parameters in Python, but we believe Matlab would be more intuitive
%%  清空环境变量
warning off             % 关闭报警信息
close all               % 关闭开启的图窗
clear                   % 清空变量
clc                     % 清空命令行

%%  导入数据
filename = 'D:\coffee.xlsx';
data = xlsread(filename);
x = data(2:end,2:end);
y = data(2:end,1);
% 存储30次的error1和error2
errors1 = zeros(10, 1);
errors2 = zeros(10, 1);
for i = 1:10
    %%  KS法划分训练集和测试集
    ratio=0.8;
    [mx,nx]=size(x);
    mtrain=ceil(mx*ratio);
    mtest=mx-mtrain;
    c = cvpartition(y, 'HoldOut', 0.2); 
    trainIdx = training(c);
    testIdx = test(c);
    P_train = x(trainIdx, :); 
    T_train = y(trainIdx, :); 
    P_test = x(testIdx, :); 
    T_test = y(testIdx, :); 
    P_train = P_train';
    P_test = P_test';
    T_train = T_train';
    T_test = T_test';
    M = size(P_train, 2);
    N = size(P_test, 2);
    
    %%  数据归一化
    [P_train, ps_input] = mapminmax(P_train, 0, 1);
    P_test  = mapminmax('apply', P_test, ps_input);
    
    t_train =  categorical(T_train)';
    t_test  =  categorical(T_test )';
    
    %%  数据平铺
    p_train =  double(reshape(P_train, 601, 1, 1, M)); 
    p_test  =  double(reshape(P_test , 601, 1, 1, N));
    
    %%  构造网络结构
    layers = [
     imageInputLayer([601,1,1])             
    
     convolution2dLayer([4,1], 16,'Stride' , [1,1], 'Padding','same')          
     batchNormalizationLayer                 
     reluLayer                               
     
     convolution2dLayer([4,1], 32,'Stride' , [1,1], 'Padding','same')          
     batchNormalizationLayer                 
     reluLayer                               
     maxPooling2dLayer([4,1], 'Stride' , [2,1])   
     fullyConnectedLayer(7)                  
     softmaxLayer                            
     classificationLayer];                   
    
    %%  参数设置
    options = trainingOptions('adam', ...      
        'MaxEpochs', 100, ...                  
        'MiniBatchSize', 28, ...                  
        'InitialLearnRate', 1e-3, ...         
        'L2Regularization', 1e-2, ...         
        'LearnRateSchedule', 'piecewise', ...  
        'LearnRateDropFactor', 0.5, ...        
        'LearnRateDropPeriod', 100, ...         
        'Shuffle', 'every-epoch', ...          
        'ValidationPatience', Inf, ...         
        'Plots', 'training-progress', ...      
        'Verbose', false);
    
    
    %%  训练模型
    net = trainNetwork(p_train, t_train, layers, options);
    
    %%  预测模型
    t_sim1 = predict(net, p_train); 
    t_sim2 = predict(net, p_test ); 
    
    %%  反归一化
    T_sim1 = vec2ind(t_sim1');
    T_sim2 = vec2ind(t_sim2');
    
    %%  性能评价
    errors1(i) = sum((T_sim1 == T_train)) / M * 100 ;
    errors2(i) = sum((T_sim2 == T_test )) / N * 100 ;
end

% 打印30次的error1和error2
disp('30次的error1和error2：');
disp('error1:');
disp(errors1);
disp('error2:');
disp(errors2);

% 计算30次的平均值
avg_error1 = mean(errors1);
avg_error2 = mean(errors2);

disp(['30次的平均error1: ', num2str(avg_error1)]);
disp(['30次的平均error2: ', num2str(avg_error2)])


##################################################################################################
# Author：Yun Zheng
# Email：18848234471@163.com
# If there are any shortcomings, please forgive me,because this is my first time sharing.
##################################################################################################
# Model - CNN--Grape - matlab(Y、Y-B、Y-B-G)
# It can also be run with the same parameters in Python, but we believe Matlab would be more intuitive
%%  清空环境变量
warning off             % 关闭报警信息
close all               % 关闭开启的图窗
clear                   % 清空变量
clc                     % 清空命令行

%%  导入数据
filename = 'D:\putao.xlsx';
data = xlsread(filename);
x = data(2:end,2:end);
y = data(2:end,1);
% 存储30次的error1和error2
errors1 = zeros(30, 1);
errors2 = zeros(30, 1);
for i = 1:30
    %%  KS法划分训练集和测试集
    ratio=0.8;
    [mx,nx]=size(x);
    mtrain=ceil(mx*ratio);
    mtest=mx-mtrain;
    c = cvpartition(y, 'HoldOut', 0.2); 
    trainIdx = training(c);
    testIdx = test(c);
    P_train = x(trainIdx, :); 
    T_train = y(trainIdx, :); 
    P_test = x(testIdx, :);
    T_test = y(testIdx, :); 
    P_train = P_train';
    P_test = P_test';
    T_train = T_train';
    T_test = T_test';
    M = size(P_train, 2);
    N = size(P_test, 2);
    
    %%  数据归一化
    [P_train, ps_input] = mapminmax(P_train, 0, 1);
    P_test  = mapminmax('apply', P_test, ps_input);
    
    t_train =  categorical(T_train)';
    t_test  =  categorical(T_test )';
    
    %%  数据平铺
    p_train =  double(reshape(P_train, 100, 1, 1, M)); 
    p_test  =  double(reshape(P_test , 100, 1, 1, N));
    
    %%  构造网络结构
    layers = [
     imageInputLayer([100,1,1])             
    
     convolution2dLayer([4,1], 16,'Stride' , [1,1], 'Padding','same')          
     batchNormalizationLayer                 
     reluLayer                               
     
     convolution2dLayer([4,1], 32,'Stride' , [1,1], 'Padding','same')         
     batchNormalizationLayer                 
     reluLayer                              
     maxPooling2dLayer([4,1], 'Stride' , [2,1])  
    
     fullyConnectedLayer(20)                   
     softmaxLayer                           
     classificationLayer];                 
    
    %%  参数设置
    options = trainingOptions('adam', ...      
        'MaxEpochs', 100, ...                 
        'MiniBatchSize', 160, ...                 
        'InitialLearnRate', 1e-3, ...         
        'L2Regularization', 1e-2, ...         
        'LearnRateSchedule', 'piecewise', ...  
        'LearnRateDropFactor', 0.5, ...        
        'LearnRateDropPeriod', 100, ...        
        'Shuffle', 'every-epoch', ...          
        'ValidationPatience', Inf, ...         
        'Plots', 'training-progress', ...      
        'Verbose', false);
    
    
    %%  训练模型
    net = trainNetwork(p_train, t_train, layers, options);
    
    %%  预测模型
    t_sim1 = predict(net, p_train); 
    t_sim2 = predict(net, p_test ); 
    
    %%  反归一化
    T_sim1 = vec2ind(t_sim1');
    T_sim2 = vec2ind(t_sim2');
    
    %%  性能评价
    errors1(i) = sum((T_sim1 == T_train)) / M * 100 ;
    errors2(i) = sum((T_sim2 == T_test )) / N * 100 ;
end

% 打印30次的error1和error2
disp('30次的error1和error2：');
disp('error1:');
disp(errors1);
disp('error2:');
disp(errors2);

% 计算30次的平均值
avg_error1 = mean(errors1);
avg_error2 = mean(errors2);

disp(['30次的平均error1: ', num2str(avg_error1)]);
disp(['30次的平均error2: ', num2str(avg_error2)])
```
